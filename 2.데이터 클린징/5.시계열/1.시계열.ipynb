### 시계열 데이터
* 시계열 데이터
    - 시간의 흐름에 따라 데이터의 변화를 분석 및 예측하는데 사용되는 데이터
    - 타임스탬프(timestamp) : 특정 시점(시간)을 의미하는 자료형
* 기능
    - to_datetime : 데이터를 시간 자료형으로 변환시켜주는 함수
    - to_period : 날짜 데이터의 년, 월, 일 등을 가져올 수 있다.
        * 옵션
            - freq를 이용하여 각각의 날짜 정보를 얻어올 수 있다
            - freq = A(년), M(월), D(일)
            - D지정시 2020-01-01까지 나오며, M지정시 2020-01, A지정시 2020까지 나온다.
    - Datetimeindex자료형
        * dt객체를 제공하며, dt.year, dt.month, dt.day, dt.hour, dt.minute, dt.second...등을 제공한다
>
* 자세한 정보는 다음 사이트를 참조하기 바란다.
* https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DatetimeIndex.html

```python
import pandas as pd
dates = ['2026-03-04', '2026-05-05']
df = pd.DataFrame(dates, columns=['date'])
df
```

```python
df['new_date'] = pd.to_datetime(df['date']) # 데이터를 시간자료형으로 변환환
df.info()
```

```python
print(df['new_date'].dt.year)
df['year'] = df['new_date'].dt.year
df['month'] = df['new_date'].dt.month
df['day'] = df['new_date'].dt.day

df
```

```python
dates = ['2026-03-04', '2026-05-05']
df = pd.DataFrame(dates, columns=['date'])
df
```

```python
datetimer = df['date'].apply(lambda x : pd.to_datetime(x))
datetimer.info(), datetimer
```

```python
df['year']=datetimer.apply(lambda x : x.year)
df['month']=datetimer.apply(lambda x : x.month)
df['day']=datetimer.apply(lambda x : x.day)
df['dayofweek']=datetimer.apply(lambda x : x.dayofweek) #dayofweek: 요일
df
```

```python
df['month'] = df['month'].apply(lambda x : 1111 if x > 3 else x)
df
```

```python
mask = (df['year'] == 2026) & (df['month'] == 1111)
df[(df['year'] == 2026) & (df['month'] == 1111)]
```

```python
print(df[mask])
print("-"*20)
df.loc[mask, 'dayofweek'] = df[mask]['dayofweek'].apply(lambda x : 111 if x == 1 else x)
df
```

```python
df_test = df[['date','year']]
df_test
```

```python
df_test = df_test.to_dict('list') # list 형식으로 만듬
df_test
```

```python
df_date = df_test['date']
print(df_date)
df_year = df_test['year']
print(df_year)
```
