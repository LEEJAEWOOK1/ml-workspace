```python
import pandas as pd
```

```python
df = pd.read_csv("../../data_set/2.데이터 클린징/loan_train_quiz_cleaning.csv")
df.head(2)
```

```python
df.columns
```

```python
features = ['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area'] # 입력 데이터(문제)
label = 'Loan_Status' # 정답
```

```python
X, y = df[features],df[label]
X # y: 정답
```

```python
df.shape
```

```python
# 데이터를 “학습용 / 테스트용”으로 나누는 함수 가져온다.
from sklearn.model_selection import train_test_split
```

```python
# 80%는 학습, 20% 테스트로 사용
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape, X_test.shape, len(y_train), len(y_test)
```

### 모델 생성 및 학습
* 모델 생성 : 변수 = DecisionTreeClassifier()
* 학습 : 변수.fit(X_train, y_train)
    - 문제와 정답을 알고리즘에 학습
* 예측 : 변수.predict(X_test)
    - 확인용 문제를 predict에 넣으면 정답을 알려준다

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier() # 모델 생성
dtc.fit(X_train, y_train) #학습
```

```python
pred = dtc.predict(X_test) # 20% 학습하지 않은 데이터. 예상 답안 뽑음
pred # 예측값
```

```python
print(y_test) # 실제 정답
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred) # 정확도(얼마나 정확하게 맞췄나)
```
